```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
from keras.preprocessing import image
```

# Create categories as with_mask and without_mask

```python
categories=['with_mask','without_mask']
```

```python
data=[]
for category in categories:
    path=os.path.join('Images/train',category)
    label=categories.index(category)
    
    for file in os.listdir(path):
        img_path=os.path.join(path,file)
        img=cv2.imread(img_path)
        img=cv2.resize(img,(224,224))
        
        data.append([img,label])
```

```python
len(data)
```

## Shuffle data

```python
import random
random.shuffle(data)
```

# Seperate features and lables

```python
x=[]
y=[]
for features,label in data:
    x.append(features)
    y.append(label)
```

```python
len(x),len(y)
```

```python
import numpy as np
```

```python
x=np.array(x)
y=np.array(y)
```

```python
x.shape
```

```python
y.shape
```

```python
y[1]
```

## Normalize features

```python
x=x/255
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

### Import VGG16 model

```python
from keras.applications.vgg16 import VGG16
```

```python
vgg=VGG16()
```

```python
vgg.summary()
```

```python
from keras import Sequential
```

# Create Model

```python
model=Sequential()
```

```python
# add all layers of VGG16 except last prediction layer
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.summary()
```

```python
from keras.layers import Dense
```

## Add last layer to model

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

# Fit Model

```python
model.fit(X_train,y_train,epochs=7,validation_data=(X_test,y_test))
```

```python
cap=cv2.VideoCapture(0)
```

```python
haar=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

### Face detection using HaarCascade

```python
def detect_face(img):
    cords=haar.detectMultiScale(img,scaleFactor=1.3, minNeighbors=2)
    return(cords)
    
```

### Face Mask Prediction

```python
def detect_face_mask(img):
    y_pred=model.predict(img.reshape(1,224,224,3))
    return y_pred
```

### Write Text on image

```python
def draw_label(img,text,pos,bg_color):
    text_size=cv2.getTextSize(text,cv2.FONT_HERSHEY_PLAIN,1,cv2.FILLED)
    end_x=pos[0]+text_size[0][0]+2
    end_y=pos[1]+text_size[0][1]-2
    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,cv2.FONT_HERSHEY_PLAIN,1,(0,0,255),1,cv2.LINE_AA)
```

## Face Mask Prediction and face Detection in Video

```python
while True:
    ret,frame=cap.read()
    #call detection method
    img=cv2.resize(frame,(224,224))
    y_pred=detect_face_mask(img)
    y_pred=y_pred[0][0]
    
    cords=detect_face(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY))
    
    for x,y,w,h in cords:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),3)
    
    if y_pred>0.5:
        y_pred=1
    else:
        y_pred=0
    
    if y_pred==0:
        draw_label(frame,'Face Mask detected',(30,30),(0,255,0))
    else:
        draw_label(frame,'No Face Mask',(30,30),(255,0,0))
    
    cv2.imshow('window',frame)
    
    if cv2.waitKey(1) & 0xFF==ord('x'):
        break
cv2.destroyAllWindows()
```
